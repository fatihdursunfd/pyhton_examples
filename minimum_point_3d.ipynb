```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import sys
```

```python
def f(x,y):
    return 0.02  * ( (0.8 * x -0.5)**2  + (0.6 * y + 0.2)**2 )
```

```python
x = np.linspace(-30, 30, 300)
y = np.linspace(-30, 30, 300)

X, Y = np.meshgrid(x, y)
Z = f(X,Y)
```

```python
x_min = -30
y_min = -30
z_min = sys.float_info.max
```

```python
for xi in x:
    for yi in y :
        z = f(xi,yi)
        if z < z_min:
            z_min = z
            x_min = xi
            y_min = yi
```

```python
print("x : ",format(x_min,".2f") , " y : ",format(y_min,".2f")," z : ",format(z_min,".2f"))
```

```python
fig = plt.figure()
ax = plt.gca(projection="3d")

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='winter', edgecolor='none')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
